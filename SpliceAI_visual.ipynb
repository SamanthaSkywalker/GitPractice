 <details>
  <summary><u>What is SpliceAI-visual?</u>...</summary>

&nbsp; &nbsp; &nbsp; &nbsp;  SpliceAI-visual is a [colab](https://colab.research.google.com/?utm_source=scs-index)-based tool which facilitates the interpretation of splicing predictions from SpliceAI ([Jaganathan et al, Cell 2019](https://www.cell.com/cell/fulltext/S0092-8674(18)31629-5)). It delivers absolute scores from SpliceAI, based on the original model, with a `10 000 bp` context window for analysis ([custom sequence function](https://github.com/Illumina/SpliceAI)).
<br/> </details>

 <details>
  <summary><u>Why do I need SpliceAI-visual?</u>...</summary>

&nbsp; &nbsp; &nbsp; &nbsp; SpliceAI is currently one of the most performant splicing predictor. But the interpretation of its delta scores can be tricky. <br/> &nbsp; &nbsp; &nbsp; &nbsp; For example, go check the delta scores of this *KMT2D* variant: [c.5782+1G>A](https://spliceailookup.broadinstitute.org/#variant=12-49042740-C-T&hg=38&distance=50&mask=0&precomputed=1). This variant is found in several [heterozygous carriers in gnomAD](https://gnomad.broadinstitute.org/variant/12-49042740-C-T?dataset=gnomad_r3), which seems at odds with the severity of *KMT2D* related phenotype. Interestingly, this variant is predicted by SpliceAI to have a in-frame rescuing donor site, but with a mild delta score of `+0.28`. Actually, using **SpliceAI-visual**, you will see that this rescuing donor site achieves an absolute score of `>0.99`, which is far from mild. How is this possible? Well, it's simply because the `+0.28` is adding up from a non-zero value of the reference allele probability (`0.71`). So, as 1 is the maximum probability, this score couldn't go up much higher.
<br/> &nbsp; &nbsp; &nbsp; &nbsp; Also, the standard version of SpliceAI currently doesn't annotate multinucleotide deletion-insertion like AG>TC. **SpliceAI-visual** does. Please, read [our article](https://doi.org/10.1186/s40246-023-00451-1) to learn more<br/></details>

<details>
  <summary><u>Which transcript should I use for my variant & does it really matter?</u>...</summary>

&nbsp; &nbsp; &nbsp; &nbsp; Yes, it matters. Especially if your variant is near the start or the end of the gene. As transcripts UTRs can vary significantly, and as SpliceAI is sensitive to long range context, be sure to use a good transcript. By default, consider using the [MANE](https://www.ensembl.org/info/genome/genebuild/mane.html) transcript, when available.<br/>
*Note: the easiest way to find the MANE transcript of your gene is by opening [gnomAD](https://gnomad.broadinstitute.org/), then select v3, enter your gene and get the MANE transcript.
If you wish to know more about the selection of transcript starting and ending, visit [this page](https://www.ncbi.nlm.nih.gov/refseq/MANE/#accessing-mane-select-data).* <br/> </details>

```
#@markdown <font color='grey'> this cell builds the sequences (full gene ref and alt sequences)
# build sequences to analyze
if ENST =='':
  from IPython.display import Markdown, display
  def printmd(string):
      display(Markdown(string))
  printmd("⚠️ **You must enter a <font color='red'>transcript</font> or a <font color='red'>gene</font> in the `ENST` field**")
  raise SystemExit()

CDNA = requests.get('http://rest.ensembl.org/sequence/id/'+ENST+'?', headers={ "Content-Type" : "application/json"})
data = CDNA.json()

def reverse_complement(seq):
  complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
  return "".join(complement.get(base, base) for base in reversed(seq))

def build_mut_seq(data, var, pad=None):
  """Returns the alternative sequence
  Arguments:
   data: dictionary which contains the data of the ref seq
   var: variant. Example: '2-240778037-CTCACCA-TTCCT'
   pad: padding. If not provided, returns the whole sequence,
     otherwise returns the seq between pos - n and pos + n

  Raises:
   ValueError if strand is not 1 or -1
   ValueError if ref(given) is different than ref found in seq from ensembl
  """
  _, pos, ref, alt = var.split('-')
  _, _, _, start, end, strand = data['desc'].split(':')
  pos = int(pos)
  strand  = int(strand)
  start = int(start)
  end = int(end)

  if not start < pos < end:
    raise ValueError(
        f"The position given (hg38)g.{pos} was not found in {ENST}.\n"
        f"Please check your ENST and your variant hg38 coordinate."
        )

  if strand == 1:
    if data['seq'][pos - start: pos - start + len(ref)] != ref:
      raise ValueError(
          f"Please check the hg38 coordinate of your variant.\n"
          f"The reference allele should be ({data['seq'][pos - start: pos - start + len(ref)]}) instead of ({ref})."
          )
    if pad:
      index_min = max(0, pos - start - pad)
      index_max = min(len(data['seq']), pos - start - len(ref) + pad)
    else:
      index_min = 0
      index_max = len(data['seq'])
    return data['seq'][index_min:pos - start] + alt + data['seq'][pos - start + len(ref):index_max]

  if strand == -1:
    if reverse_complement(data['seq'][end - pos - len(ref) + 1 :end - pos + 1]) != ref:
      raise ValueError(
          f"Please check the hg38 coordinate of your variant.\n"
          f"The reference allele should be ({reverse_complement(data['seq'][end - pos - len(ref) + 1:end - pos + 1])}) instead of ({ref})."
          )
    if pad:
      index_min = max(0, end - pos - len(ref) + 1 - pad)
      index_max = min(len(data['seq']), end - pos + 1 + pad)
    else:
      index_min = 0
      index_max = len(data['seq'])
    return data['seq'][index_min:end - pos - len(ref) + 1] + reverse_complement(alt) + data['seq'][end - pos + 1:index_max]
  raise ValueError(f'Strand = {strand}')

#data = load_seq(ENST)

_, _, _, start, end, strand = data['desc'].split(':')
start = int(start)
end = int(end)
strand  = int(strand)

_, pos, ref, alt = var.split('-')
pos = int(pos)

len(data['seq'])

WT_sequence = data['seq']
mut_sequence = build_mut_seq(data, var)
```

```
#@markdown <font color='grey'> this cell runs SpliceAI
!pip install spliceai
!pip install tensorflow

from keras.models import load_model
from pkg_resources import resource_filename
from spliceai.utils import one_hot_encode
import numpy as np

context = 10000
paths = ('models/spliceai{}.h5'.format(x) for x in range(1, 6))
models = [load_model(resource_filename('spliceai', x)) for x in paths]

x = one_hot_encode('N'*(context//2) + WT_sequence + 'N'*(context//2))[None, :]
y = np.mean([models[m].predict(x) for m in range(5)], axis=0)

wt_acceptor_prob = y[0, :, 1]
wt_donor_prob = y[0, :, 2]

x = one_hot_encode('N'*(context//2) + mut_sequence + 'N'*(context//2))[None, :]
z = np.mean([models[m].predict(x) for m in range(5)], axis=0)

mut_acceptor_prob = z[0, :, 1]
mut_donor_prob = z[0, :, 2]
```
